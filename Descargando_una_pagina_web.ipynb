```python
def obtener_notas(soup):
    lista_notas = []
    
    #Articulo promocionado
    featured_article = soup.find('div',attrs={'class':'article-item__content'})
    if featured_article:
        lista_notas.append(featured_article.a.get('href'))
        
    #Listado de articulos TOP
    article_list_top = soup.find('section', attrs = {'class':'list-content'})
    for article in article_list_top.find_all('h4'):
        if article.a:
            lista_notas.append(article.a.get('href'))
    
    #Listado de articulos SUB
    #article_list_sub = s_section.find('section', attrs = {'class':'list-content'}).find_all('h4')
    #for article in article_list_sub:
       # if article.a:
        #   lista_notas.append(article.a.get('href'))
    
    return lista_notas
    
```

```python
url_nota = lista_notas[0]
url_nota
```

```python
try:
    nota = requests.get(url_nota)
    if nota:
        s_nota = BeautifulSoup(nota.text, 'lxml')
        
        # Titulo
        titulo = s_nota.find('h1', attrs = {'class':'article-title'})
        print(titulo.text)
        print('\n')
        
        # Fecha
        fecha = s_nota.find('span', attrs={'pubdate': 'pubdate'}).get('datetime')
        print(fecha)
        print('\n')
        
        #Volanta
        volanta = s_nota.find('h2', attrs={'class':'article-prefix'})
        print(volanta.text)
        print('\n')
        
        #copete
        try:
            
            copete = s_nota.find('div', attrs={'class': 'article-summary'})
            print(copete.text)
            print('\n')
            
        except:
            print(None)
            print('\n')
        
        #autor
        autor = s_nota.find('div', attrs={'class':'article-author'}).a
        print(autor.text)
        print('\n')
        
        #texto
        texto = s_nota.find('div', attrs={'class':'article-text'}).find_all('p')
        articulo_texto = ''
        for pedazo_texto in texto:
            articulo_texto += pedazo_texto.text
        
        print(articulo_texto)
        
except Exception as e:
    print('Error')
    print(e)
    print('*'*50)
    print('\n')
```
